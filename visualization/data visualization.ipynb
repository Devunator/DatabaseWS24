```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
Q1 = pd.read_csv('C:/Users/pc/OneDrive/Desktop/Q1.csv')
print(Q1)
```

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.barplot(x='total_gross', y='genre', data=Q1, palette='viridis')

plt.xlabel('Total Gross (in billions)', fontsize=12)
plt.ylabel('Genre', fontsize=12)
plt.title('Most Successful Genres of Movies', fontsize=15)

plt.tight_layout()
plt.show()
```

```python
Q2 =  pd.read_csv('C:/Users/pc/OneDrive/Desktop/Q2.csv')
print(Q2)
```

```python
sns.barplot(
    data=Q2, 
    x='average_gross', 
    y='companyname', 
    hue='genre', 
    dodge=False,  
    palette='viridis'
)

plt.xlabel('Average Gross (in billions)', fontsize=12)
plt.ylabel('Company Name', fontsize=12)
plt.title('Most Successful Film Studios and Their Most Successful Genre', fontsize=15)

plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```

```python
Q3 = pd.read_csv('C:/Users/pc/OneDrive/Desktop/Q3.csv')
print(Q3)
```

```python
correlation = Q3['gross'].corr(Q3['imdbscore'])
print(f"Correlation between Gross and IMDb Score for Top 100 Movies: {correlation:.2f}")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=Q3, x='gross', y='imdbscore', color='purple', s=100)

plt.xlabel('Gross (in billions)', fontsize=12)
plt.ylabel('IMDB Score', fontsize=12)
plt.title('Correlation between Movie Gross (Success) and IMDb Score (Popularity)', fontsize=15)

plt.tight_layout()
plt.show()
```

```python
Q5 = pd.read_csv('C:/Users/pc/OneDrive/Desktop/Q5.csv')
print(Q5)
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=Q5, x='total_gross', y='starname', palette='viridis')

plt.xlabel('Total Gross (in billions)', fontsize=12)
plt.ylabel('Actor Name', fontsize=12)
plt.title('Most Profitable Actors by Total Gross', fontsize=15)

plt.tight_layout()
plt.show()

```

```python
genre_data = Q5.pivot_table(values='genre_count', index='starname', columns='genre', aggfunc='sum', fill_value=0)

# Plot the stacked bar chart
genre_data.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='Set3')

# Adding labels and title for Genre Distribution plot
plt.xlabel('Actor Name', fontsize=12)
plt.ylabel('Genre Count', fontsize=12)
plt.title('Genre Distribution for Most Profitable Actors', fontsize=15)

# Show the plot for Genre Distribution
plt.tight_layout()
plt.show()
```
